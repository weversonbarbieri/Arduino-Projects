```python
# Driver list
bott = 'bottas'
verstappen = 'max_verstappen'
norr = 'norris'
albo = 'albon'
hami = 'hamilton'
```

```python
# This function check the name of each driver and return a string with the full name of the driver.
def drivers(x):
  if x in bott:
    return 'Valtteri Bottas'
  elif x in verstappen:
    return 'Max Verstappen'
  elif x in norr:
    return 'Lando Norris'
```

```python
# Create a column with the name "Driver_Name" and apply the function "drivers" to the dataframe:
df['Driver_Name'] = df['Driver_Id'].apply(drivers)
```

```python
mercedes = ['Valtteri Bottas']
bed_bull = ['Max Verstappen']
mcLaren = ['Lando Norris']
```

```python
def F1team(x):
  if x in mercedes:
    return 'Mercedes'
  elif x in bed_bull:
    return 'Red Bull'
  elif x in mcLaren:
    return 'McLaren'
```

```python
df['Driver_Team'] = df['Driver_Name'].apply(F1team)
```
